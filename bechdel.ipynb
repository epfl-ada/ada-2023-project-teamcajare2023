```python
# convert year (float) in string so that it can be join on it 
df_bechdel['year'] = df_bechdel['year'].astype(str)
df_bechdel.dtypes
```

```python
def reformat_title(title):
    # Ensure title is a string and not None or NaN
    if isinstance(title, str):
        # Check if the title ends with ', the' (ignoring case)
        if title.lower().endswith(', the'):
            # Remove the last 5 characters (', the') and strip any leading/trailing whitespace
            new_title = title[:-5].strip()
            # Return the reformatted title with 'The ' at the beginning
            return 'the ' + new_title
        else:
            # If the title does not end with ', the', return it as is
            return title
    else:
        return title

# Apply the function to the 'title' column
df_bechdel['title'] = df_bechdel['title'].apply(reformat_title)
df_bechdel.iloc[412]
```

```python
# merge both datasets on title and year to avoid duplicates in title 
df_merged = df_movies.merge(df_bechdel, on=('title', 'year'), how='left')
df_merged
```

```python
# Percentage of films with rating available 
count_per_colums = df_merged.count()
percentage_movies_rate = count_per_colums['rating'] / count_per_colums['title'] * 100
print("Percentage of film with rating available is", round(percentage_movies_rate, 2), "%")

# Is it too small to be used? 
# do we need to merge both dataset?
```
