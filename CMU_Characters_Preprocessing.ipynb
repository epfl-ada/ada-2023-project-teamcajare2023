## Release Date

We extract the release year of the movies, and we convert the release date in format datetime.

```python
def convert_date(date_string):
    try:
        return pd.to_datetime(date_string, format='%Y-%m-%d')
    except ValueError:
        try:
            return pd.to_datetime(date_string, format='%Y-%m')
        except ValueError:
            try:
                return pd.to_datetime(date_string, format='%Y')
            except ValueError:
                try: 
                    return pd.to_datetime(date_string, format='%Y-%m-%d %H:%M:%S.%f')
                except ValueError:
                    return np.nan


#There is a movie that has incorrect date. It is named 'Hunting Season' and was released in 2010,
# but it is written that it was released in 1010. So we need to correct this.
df_characters.loc[df_characters['ReleaseDate']=='1010-12-02']
df_characters['ReleaseDate'] = df_characters['ReleaseDate'].replace('1010-12-02', '2010-12-02')      

# Convert the date in the column 'ReleaseDate' into datetime
df_characters['ReleaseDate'] = df_characters['ReleaseDate'].apply(convert_date)
ReleaseDate_datatime = df_characters['ReleaseDate']

# Add a column with the year of release only
df_characters['ReleaseYear'] = ReleaseDate_datatime.dt.year.fillna(0).astype(int) #OKAY TO HAVE NAN AS 0 ?
df_characters['ReleaseYear']=df_characters['ReleaseYear'].replace(0, np.nan)
```

## Age

Estimated age of the actor at the date of movie release is calculated from its date of birth and the date of movie release.

```python
df_characters['Age'].describe()
```

The smallest estimated age in the dataset is negative, which is impossible. We then take a deeper look into the negative and null values of this column.

```python
display(df_characters[df_characters['Age'] <= 0])
```

421 characters (0.01% of the dataset) have negative age; we choose to replace their age by NaN. 

```python
# Replace negative and null values of the dataset by NaN
df_characters['Age'] = df_characters['Age'].apply(lambda x: np.nan if x <= 0 else x)
```

## Date of birth

We first convert the date of birth into datetime data type. Then we extract the year of birth, as we don't care about the complete date.

```python
df_characters['DoB'] = df_characters['DoB'].apply(convert_date)
df_characters['DoB'] = df_characters['DoB'].dt.year
df_characters.rename(columns = {"DoB": "YoB"}, inplace=True)
```

```python
display(df_characters)
```

```python
boxplot = df_characters.boxplot(column=['YoB'])
```

Our last release date is 2016; thus we replace any year of birth & date of birth above 2015 by NaN.

```python
df_characters['YoB'] = df_characters['YoB'].apply(lambda x: np.nan if x >= 2016 else x)
```

```python
women_characters = df_characters[df_characters['gender'] == 'F']
men_characters = df_characters[df_characters['gender'] == 'M']

number_of_women_by_movie = women_characters.groupby('FreebaseMovieId')['character name'].nunique()
number_of_men_by_movie = men_characters.groupby('FreebaseMovieId')['character name'].nunique()

total_character_by_movie = df_characters.groupby('FreebaseMovieId')['character name'].nunique()

percentage_of_women_by_movie = number_of_women_by_movie / total_character_by_movie * 100
percentage_of_men_by_movie = number_of_men_by_movie / total_character_by_movie * 100

plt.hist(percentage_of_men_by_movie, bins=50, label='men', alpha=0.5)
plt.hist(percentage_of_women_by_movie, bins=50, label='women', alpha=0.5)
plt.legend(loc='upper right') 
plt.show()
```
